```python
import ast
def convert_column(txt):
    lis = []
    for i in ast.literal_eval(txt):
        lis.append(i['name'])
    return lis     
```

```python
# Apply the 'convert_column' function to the 'genres' column of the 'movies' DataFrame
movies['genres'] = movies['genres'].apply(convert_column)
```

```python
movies.head(2)
```

```python
# Apply the 'convert_column' function to the 'keywords' column of the 'movies' DataFrame
movies['keywords'] = movies['keywords'].apply(convert_column)
```

```python
movies.head(2)
```

```python
#from cast column, we are fetching only top 3 actors name(actual name of a actor, main lead actors) for the respective movie
def first_3top_actor(txt):
    lis = []
    count = 0
    for i in ast.literal_eval(txt):
        if count < 3:
            lis.append(i['name'])
        count+=1
    return lis
```

```python
# Apply the 'first_3top_actor' function to the 'cast' column of the 'movies' DataFrame
movies['cast'] = movies['cast'].apply(first_3top_actor)
```

```python
movies.head(2)
```

```python
movies['overview'][0]
```

```python
#Need to convert overview column into list for concatenation as it is in string and list+string concatenation not possible
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head(2)
```

```python
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(' ','')for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(' ','')for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(' ','')for i in x])
```

```python
movies.head(3)
```

```python
movies['tags'] =movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast']
```

```python
Df = movies[['movie_id','title','tags']]
```

```python
Df
```

```python
Df['tags'] = Df['tags'].apply(lambda x: ' '.join(x))
```

```python
Df.head()
```

```python
Df['tags'][0]
```

```python
# converting in lower case of all the texes
```

```python
Df['tags'] = Df['tags'].apply(lambda x: x.lower())
```

```python
Df.head(88)
```

```python
# steaming words 
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(txt):
    t = []
    for i in txt.split():
        t.append(ps.stem(i))
    return ''.join(t)  
```

```python
Df['tags'] = Df['tags'].apply(stem)
```

```python
Df
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector = cv.fit_transform(Df['tags']).toarray()
```

```python
vector.shape
```

```python
cv.get_feature_names()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vector)
```

```python
similarity
```

```python
def recommended_movie(movie):
    m_index = Df[Df['title'] == movie].index[0]
    dist = similarity[m_index]
    mo_list = sorted(list(enumerate(dist)), reverse=True, key=lambda x: x[1])[1:6]
    for i in mo_list:
        print(Df.iloc[i[0]]['title'])
```

```python
def recommended_moviee(movie):
    movie_index = Df[Df['title'] == movie].index
    if len(movie_index) == 0:
        print("Movie not found.")
        return
    m_index = movie_index[0]
    if m_index >= len(similarity):
        print("Invalid movie index.")
        return
    dist = similarity[m_index]
    mo_list = sorted(list(enumerate(dist)), reverse=True, key=lambda x: x[1])[1:11]
    for i in mo_list:
        if i[0] >= len(Df):
            continue
        print(Df.iloc[i[0]]['title'])

```

```python
# this Is How Showing Top 10 Similar movie 
recommended_moviee('Avatar')
```
